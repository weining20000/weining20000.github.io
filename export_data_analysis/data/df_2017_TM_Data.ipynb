```python
import pandas as pd
```

```python
df = pd.read_excel('2017_Data_xl_version.xlsx', header=2, index_col=None)
df.head(5)
```

```python
list(df)
```

```python
df1 = df[["Reporting Economy", "Product/Sector", "Partner Economy", "2017"]]
df1.head()
```

```python
#Asia = df1.loc[(df1['Reporting Economy'] == "Asia") & (df1['Product/Sector'] == "SI3_AGG - TO - Total merchandise")]
#Asia
```

```python
Asia = df1.loc[(df1['Reporting Economy'] == "Asia") & (df1['Product/Sector'] == "SI3_AGG - TO - Total merchandise")]
dfAsia = Asia[Asia['Partner Economy'].isin(["Asia", 'Middle East', 'Africa', 'South and Central America and the Caribbean', 'North America', 'Europe', 'Commonwealth of Independent States (CIS), including associate and former member States'])]  
dfAsia
```

```python
ME = df1.loc[(df1['Reporting Economy'] == "Middle East") & (df1['Product/Sector'] == "SI3_AGG - TO - Total merchandise")]
dfME = ME[ME['Partner Economy'].isin(['Asia', "Middle East", 'Africa', 'South and Central America and the Caribbean', 'North America', 'Europe', 'Commonwealth of Independent States (CIS), including associate and former member States'])]  
dfME
```

```python
Africa = df1.loc[(df1['Reporting Economy'] == "Africa") & (df1['Product/Sector'] == "SI3_AGG - TO - Total merchandise")]
dfAfrica = Africa[Africa['Partner Economy'].isin(['Asia', 'Middle East', "Africa", 'South and Central America and the Caribbean', 'North America', 'Europe', 'Commonwealth of Independent States (CIS), including associate and former member States'])]  
dfAfrica
```

```python
SCA = df1.loc[(df1['Reporting Economy'] == "South and Central America and the Caribbean") & (df1['Product/Sector'] == "SI3_AGG - TO - Total merchandise")]
dfSCA = SCA[SCA['Partner Economy'].isin(['Asia', 'Middle East', 'Africa', "South and Central America and the Caribbean", 'North America', 'Europe', 'Commonwealth of Independent States (CIS), including associate and former member States'])]  
dfSCA
```

```python
NA = df1.loc[(df1['Reporting Economy'] == "North America") & (df1['Product/Sector'] == "SI3_AGG - TO - Total merchandise")]
dfNA = NA[NA['Partner Economy'].isin(['Asia', 'Middle East', 'Africa', 'South and Central America and the Caribbean', "North America", 'Europe', 'Commonwealth of Independent States (CIS), including associate and former member States'])]  
dfNA
```

```python
Europe = df1.loc[(df1['Reporting Economy'] == "Europe") & (df1['Product/Sector'] == "SI3_AGG - TO - Total merchandise")]
dfEurope = Europe[Europe['Partner Economy'].isin(['Asia', 'Middle East', 'Africa', 'South and Central America and the Caribbean', 'North America', "Europe", 'Commonwealth of Independent States (CIS), including associate and former member States'])]  
dfEurope
```

```python
CIS = df1.loc[(df1['Reporting Economy'] == "Commonwealth of Independent States (CIS), including associate and former member States") & (df1['Product/Sector'] == "SI3_AGG - TO - Total merchandise")]
dfCIS = CIS[CIS['Partner Economy'].isin(['Asia', 'Middle East', 'Africa', 'South and Central America and the Caribbean', 'North America', 'Europe', "Commonwealth of Independent States (CIS), including associate and former member States"])]  
dfCIS
```

```python
dfAll = pd.concat([dfAsia, dfME, dfAfrica, dfSCA, dfNA, dfEurope, dfCIS], ignore_index=True)
dfAll
```

```python
dfTM2017 = dfAll.drop(columns=['Product/Sector'])

#rename column names
dfTM2017 = dfTM2017.rename(columns={"Reporting Economy": "From", "Partner Economy": "To", "2017":"Export_Value"})
dfTM2017.head()
```

```python
dfTM2017_2 = dfTM2017.replace({'South and Central America and the Caribbean': "SCA", 'Commonwealth of Independent States (CIS), including associate and former member States': 'CIS'})
dfTM2017_2
```

```python
dfTM2017_2['Export_Value_Bn'] = dfTM2017_2['Export_Value'].div(1000).round(2)
dfTM2017_2.head()
```

```python
dfTM2017_3 = dfTM2017_2.drop(columns=['Export_Value'])
dfTM2017_3.head()
```

```python
dfTM2017_3.to_csv("2017_TM_Data_by_Region.csv", sep=',', encoding='utf-8', index = None, header=True)
```

```python
dfTM2017_pivot = dfTM2017_3.pivot_table('Export_Value_Bn', ['From'], 'To')
dfTM2017_pivot
```

```python
matrix = dfTM2017_pivot.values
matrix
```
